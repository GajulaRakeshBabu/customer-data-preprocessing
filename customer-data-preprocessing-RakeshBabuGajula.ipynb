```python
import numpy as np
import pandas as pd

data = {
    "customer_id": [1, 2, 2, 3, 4, 5],
    "age": [25, 30, 30, np.nan, 45, np.nan],
    "city": ["Delhi", "Mumbai", "Mumbai", "Delhi", np.nan, "Bangalore"],
    "gender": ["Male", "Female", "Female", "Male", "Female", "Male"],
    "annual_income": [50000, 60000, 60000, 55000, 65000, 70000]
}

df = pd.DataFrame(data)

age_median = df["age"].median()
df["age"] = df["age"].fillna(age_median)

city_mode = df["city"].mode()[0]
df["city"] = df["city"].fillna(city_mode)

df = df.drop_duplicates().reset_index(drop=True)

print("Cleaned DataFrame (Task 1):")
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df_encoded = pd.get_dummies(df, columns=["city"], prefix="city")

label_encoder = LabelEncoder()
df_encoded["gender"] = label_encoder.fit_transform(df_encoded["gender"])

print("DataFrame after encoding:")
print(df_encoded)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

num_cols = ["age", "annual_income"]
X_num = df[num_cols]

min_max_scaler = MinMaxScaler()
X_minmax = min_max_scaler.fit_transform(X_num)

df_minmax = pd.DataFrame(X_minmax, columns=[col + "_minmax" for col in num_cols])

print("Min-Max Scaled features:")
print(df_minmax)

standard_scaler = StandardScaler()
X_standard = standard_scaler.fit_transform(X_num)

df_standard = pd.DataFrame(X_standard, columns=[col + "_standard" for col in num_cols])

print("\nStandardised features:")
print(df_standard)

df_scaled_combined = pd.concat([df.reset_index(drop=True), df_minmax, df_standard], axis=1)

print("\nCombined DataFrame with original and scaled features:")
print(df_scaled_combined)
```
